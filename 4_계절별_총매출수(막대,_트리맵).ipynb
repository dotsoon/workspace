```python
spring1 = spring[['ANALS_YM', 'CMPNM_NM', 'SIGNGU_NM', 'ADSTRD_NM', 'AREA_NM', 'MLSFC_NM', 'SCLAS_NM',
         'JJINHBT_SALES_CO_RATE', 'OTSD_SALES_CO_RATE', 'ALL_SALES_CO_RATE'  ]]
spring_new = spring1.reset_index(drop=True)
spring_n = spring_new.rename(columns={'ANALS_YM':'날짜', 'CMPNM_NM':'상점명', 'SIGNGU_NM':'시', 'ADSTRD_NM':'읍면동', 'AREA_NM':'읍면동2', 
                      'MLSFC_NM':'업종', 'SCLAS_NM':'세부업종', 'JJINHBT_SALES_CO_RATE':'제주도민매출수', 'OTSD_SALES_CO_RATE':'외지인매출수', 
                      'ALL_SALES_CO_RATE':'총매출수'})
summer1 = summer[['ANALS_YM', 'CMPNM_NM', 'SIGNGU_NM', 'ADSTRD_NM', 'AREA_NM', 'MLSFC_NM', 'SCLAS_NM',
         'JJINHBT_SALES_CO_RATE', 'OTSD_SALES_CO_RATE', 'ALL_SALES_CO_RATE'  ]]
summer_new = summer1.reset_index(drop=True)
summer_n = summer_new.rename(columns={'ANALS_YM':'날짜', 'CMPNM_NM':'상점명', 'SIGNGU_NM':'시', 'ADSTRD_NM':'읍면동', 'AREA_NM':'읍면동2', 
                      'MLSFC_NM':'업종', 'SCLAS_NM':'세부업종', 'JJINHBT_SALES_CO_RATE':'제주도민매출수', 'OTSD_SALES_CO_RATE':'외지인매출수', 
                      'ALL_SALES_CO_RATE':'총매출수'})
fall1 = fall[['ANALS_YM', 'CMPNM_NM', 'SIGNGU_NM', 'ADSTRD_NM', 'AREA_NM', 'MLSFC_NM', 'SCLAS_NM',
         'JJINHBT_SALES_CO_RATE', 'OTSD_SALES_CO_RATE', 'ALL_SALES_CO_RATE'  ]]
fall_new = fall1.reset_index(drop=True)
fall_n = fall_new.rename(columns={'ANALS_YM':'날짜', 'CMPNM_NM':'상점명', 'SIGNGU_NM':'시', 'ADSTRD_NM':'읍면동', 'AREA_NM':'읍면동2', 
                      'MLSFC_NM':'업종', 'SCLAS_NM':'세부업종', 'JJINHBT_SALES_CO_RATE':'제주도민매출수', 'OTSD_SALES_CO_RATE':'외지인매출수', 
                      'ALL_SALES_CO_RATE':'총매출수'})
winter1 = winter[['ANALS_YM', 'CMPNM_NM', 'SIGNGU_NM', 'ADSTRD_NM', 'AREA_NM', 'MLSFC_NM', 'SCLAS_NM',
         'JJINHBT_SALES_CO_RATE', 'OTSD_SALES_CO_RATE', 'ALL_SALES_CO_RATE'  ]]
winter_new = winter1.reset_index(drop=True)
winter_n = winter_new.rename(columns={'ANALS_YM':'날짜', 'CMPNM_NM':'상점명', 'SIGNGU_NM':'시', 'ADSTRD_NM':'읍면동', 'AREA_NM':'읍면동2', 
                      'MLSFC_NM':'업종', 'SCLAS_NM':'세부업종', 'JJINHBT_SALES_CO_RATE':'제주도민매출수', 'OTSD_SALES_CO_RATE':'외지인매출수', 
                      'ALL_SALES_CO_RATE':'총매출수'})
```

```python
spring_sales = spring_n.loc[:,['세부업종','총매출수']]
spring_all = spring_sales.groupby('세부업종').sum('총매출수')
spring_all_top = spring_all.nlargest(10,'총매출수',keep='all')
summer_sales = summer_n.loc[:,['세부업종','총매출수']]
summer_all = summer_sales.groupby('세부업종').sum('총매출수')
summer_all_top = summer_all.nlargest(10,'총매출수',keep='all')
fall_sales = fall_n.loc[:,['세부업종','총매출수']]
fall_all = fall_sales.groupby('세부업종').sum('총매출수')
fall_all_top = fall_all.nlargest(10,'총매출수',keep='all')
winter_sales = winter_n.loc[:,['세부업종','총매출수']]
winter_all = winter_sales.groupby('세부업종').sum('총매출수')
winter_all_top = winter_all.nlargest(10,'총매출수',keep='all')
```
